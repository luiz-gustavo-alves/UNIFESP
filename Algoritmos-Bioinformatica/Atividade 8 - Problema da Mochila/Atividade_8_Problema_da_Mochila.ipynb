# PROBLEMA DA MOCIHLA

O problema da mochila é um problema de optimização combinatória em que é necessário preencher uma mochila com objetos de diferentes pesos e valores. O objetivo é que se preencha a mochila com o maior valor possível, não ultrapassando o peso máximo estipulado.

### Código

```python
def knapsack(max_weigth, weigth, values, length):

    memory = []

    ## Criando matriz de programacao dinamica
    for i in range(length + 1):
      memory.append([])

    ## Algoritmo para solucionar o problema da mochila
    for i in range(length + 1):

      for j in range(max_weight + 1):

        if i == 0 or j == 0:
          memory[i].append(0)

        elif weight[i-1] > j:
          memory[i].append(memory[i-1][j])

        elif i > 0 and weight[i-1] <= j:
          memory[i].append(max(memory[i-1][j], memory[i-1][j-weight[i-1]] + values[i-1]))

    ## Retornando o valor máximo da mochila
    max_value = memory[length][max_weigth]
    return max_value, memory
```

### Entrada e saída do algoritmo

```python
values = [5, 4, 7, 7]
weight = [5, 6, 8, 4]
max_weight = 13

max_value, memory = knapsack(max_weight, weight, values, len(values))
```

#### Código

```python
def knapsack_values(max_weight, weight, values, length, max_value, memory):

  res = max_value
  w = max_weight
  bag_values = []

  for i in range(length, 0, -1):

      if res <= 0:
          break

      ## Verificando se o valor do resultado está na matriz de memória.
      if res == memory[i - 1][w]:
          continue

      ## Valor fora da matriz de memória, incluir este item na mochila.
      else:
          bag_values.append(values[i - 1])
          res = res - values[i - 1]
          w = w - weight[i - 1]

  return bag_values
```

#### Valor de entrada e saída

```python
bag_values = knapsack_values(max_weight, weight, values, len(values), max_value, memory)

print(f"Valores dos itens da mochila: {bag_values}")
```
