# Identificar e analisar regiões do genoma da H3N2 que potencialmente codificam proteínas

## Criar função que facilite aplicação do algoritmo

```python
def proteinAnalysis(file_name, frame_number, content):

  RNA_Seqs = []

  for record in SeqIO.parse(f"{file_name}.fasta", "fasta"):

    RNA_Seqs.append(record.seq)

  DNA_segment = 0

  for RNA_Seq in RNA_Seqs:

    DNA_segment += 1
    protein_count = 0
    haveStopCodon = False
    protein = ""

    for seq in RNA_Seq:

      if (seq == "M" and not haveStopCodon): 
          haveStopCodon = True
          protein += seq

      if (haveStopCodon):
          protein += seq

          if (seq == "*"):
            protein_count += 1
            haveStopCodon = False
            protein_seq = Bio.Seq.Seq(protein)
            protein = ""

            id_content = "H2N2,"
            description_content = f"Frame {frame_number}, segmento {DNA_segment}, proteína {protein_count}"

            new_content = SeqRecord(
                protein_seq,
                id = id_content,
                description = description_content
            )
            content.append(new_content)
```

## Análise de todos os segmentos para todos os frames

```python
content = []

proteinAnalysis("EX3_ALL_RNA_FRAME_01", 1, content)
proteinAnalysis("EX3_ALL_RNA_FRAME_02", 2, content)
proteinAnalysis("EX3_ALL_RNA_FRAME_03", 3, content)
proteinAnalysis("EX3_ALL_RNA_FRAME_04", 4, content)
proteinAnalysis("EX3_ALL_RNA_FRAME_05", 5, content)
proteinAnalysis("EX3_ALL_RNA_FRAME_06", 6, content)

SeqIO.write(
    content,
    "EX4_ALL_RNA_PROTEINS.fasta",
    "fasta"
) 
```
