```python
# Importanto bibliotecas do Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

# Mostrar todas colunas do Dataset
pd.set_option('display.max_columns', None)
 
# Carregando o Dataset para Dataframe pandas
df = pd.read_csv("odds_copa_do_mundo_22_geral.csv")
df.rename(columns = {c: c.strip() for c in df.columns}, inplace = True)
df.set_index("SELEÇÕES", inplace = True)
df = df.sort_values(by = "GRUPO")

print(df.shape)
```

## **Análise Gráfica e Estatística dos dados (probabilidades) das Seleções da Copa do Mundo 2022**

```python
def boxPlotDados(df):

    X = df
    y = X.iloc[:, X.columns != "GRUPO"]

    g = sns.boxplot(data = y)
    g.set_xticks(range(len(y.columns.values)))
    g.set_xticklabels(['Fase de Grupos','Quartas','Semis', 'Final', 'Ganhar'])
    g.set_yticks(np.linspace(0, 100, 11))

    plt.title("BOXPLOT DAS PROBABILIDADES DE CADA SELEÇÃO DA COPA DO MUNDO 2022", fontsize = 16, fontweight = "bold")
    plt.ylabel("PROBABILIDADE (%)", fontsize = 12)
    plt.tick_params(axis = "x", labelsize = 14)
    plt.tick_params(axis = "y", labelsize = 14)

    plt.show()
```

```python
boxPlotDados(df)
```

## **Análise Gráfica de cada grupo das Seleções da Copa do Mundo 2022**

Gráfico probabilístico de cada seleção, em seu respectivo grupo, conseguir:
- Passar da Fase Grupos.
- Chegar nas Quartas de Final.
- Chegar nas Semifinais.
- Chegar na Final.
- Ganhar a Copa.


```python
def graficoProbGrupoSelecao(df):

    grupos = np.unique(df["GRUPO"])

    eixo_i = [0, 0, 1, 1, 2, 2, 3, 3]
    eixo_j = [0, 1, 0, 1, 0, 1, 0, 1]
    figure, axis = plt.subplots(4, 2, figsize=(20, 24))

    k = 0
    for grupo in grupos:
        X = df[df["GRUPO"] == grupo]
        X = X.sort_values(by = "PROB_FASE_GP", ascending = False)
        y = X.loc[:, X.columns != "GRUPO"]

        for i in range(len(y)):
            prob = list(y.iloc[i])
            axis[eixo_i[k]][eixo_j[k]].plot(y.columns.values, prob, 'o-')

        axis[eixo_i[k]][eixo_j[k]].legend(X.index, frameon = False, fontsize = 17)
        axis[eixo_i[k]][eixo_j[k]].set_title("Grupo " + grupo, fontsize = 15, fontweight="bold")
        axis[eixo_i[k]][eixo_j[k]].set_xticklabels(['Fase de Grupos','Quartas','Semis', 'Final', 'Ganhar'], fontsize = 15)
        axis[eixo_i[k]][eixo_j[k]].set_yticks(np.linspace(0, 100, 6))
        axis[eixo_i[k]][eixo_j[k]].tick_params(axis = "y", labelsize = 15)
        axis[eixo_i[k]][eixo_j[k]].set_ylabel("PROBABILIDADE (%)", fontsize = 15)

        k += 1

    plt.show()
```

```python
graficoProbGrupoSelecao(df)
```

## **Análise Gráfica das Seleções favoritas a ganhar a Copa do Mundo 2022**

- Gráfico para o Top 10 das Seleções Favoritas a ganhar a copa.
- Gráfico geral das 32 Seleções presentes na Copa.


```python
def graficoTop10(df):

    X = df.sort_values(by = "PROB_GANHAR_COPA", ascending = False)
    X = X.head(10)
    y = X["PROB_GANHAR_COPA"]

    plt.bar(y.index, y, color = 'maroon', width = 0.5, edgecolor = 'black')
    plt.ylabel("PROBABILIDADE (%)", fontsize = 14)
    plt.title("TOP 10 - SELEÇÕES FAVORITAS PARA GANHAR A COPA 2022", fontsize = 15, fontweight="bold")
    
    plt.tick_params(axis = "x", labelsize = 14)
    plt.tick_params(axis = "y", labelsize = 13)

    plt.show()
```

```python
graficoTop10(df)
```

```python
def graficoGeral(df):

    X = df.sort_values(by = "PROB_GANHAR_COPA", ascending = False)
    y = X["PROB_GANHAR_COPA"]

    plt.bar(y.index, y, color = 'mediumslateblue', width = 0.8, edgecolor = 'black')
    plt.ylabel("PROBABILIDADE (%)", fontsize = 14)
    plt.title("SELEÇÕES FAVORITAS PARA GANHAR A COPA 2022", fontsize = 15, fontweight="bold")

    plt.xticks(y.index, rotation = 270)
    plt.tick_params(axis = "x", labelsize = 13)
    plt.tick_params(axis = "y", labelsize = 13)

    plt.show()
```

```python
graficoGeral(df)
```
